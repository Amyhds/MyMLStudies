pandas는 엑셀과 같은 데이터를 손쉽게 읽고 쓸 수 있게 도와주는 라이브러리 입니다.
pandas는 다음과 같은 긴으을 위해 만들어 졌다고 합니다.

1. 축의 이름을 따라 데이터를 정렬할 수 있는 자료 구조(data structure). 다양한 소스에서 가져온 다양한 방식으로 색인된 데이터를 핸들링 가능한 기능

2. 통합된 시계열 기능

3. 시계열, 비시계열 데이터 모두 다룰 수 있는 자료 구조

4. 산술연산 및 한 축(column을 의미하는 듯)의 모든 값을 더하는 등의 축약연산이 가능

5. 누락된 데이터의 유연한 처리 기능

6. SQL 같은 관계연산 수행 기능

pandas.columns[idx] 는 데이터 프레임의 가리키고 싶은 열 위치를 입력하면 해당 위치의 열의 이름을 반환해 줍니다.

```python
df[df.columns[7]].head(5)
```

이번에는 열의 이름이 아닌 위치를 지정하여 데이터를 추출하는 방법에 대해 알아 봅시다.
pandas는 pandas.DataFrame.iloc method를 통해 데이터 프레임의 위치 좌표계로 데이터에 접근 할 수 있는 method를 제공해 줍니다.
df.iloc[가져올 행의 위치, 가져올 열의 위치]

```python
#df.iloc[가져올 행의 위치, 가져올 열의 위치]
healthy = df.iloc[:,7]
healthy.head(5)
```

그렇다면 이번에는 데이터 프레임의 열을 제거해 보는 것을 실습해 보도록 합니다.
아래 보이는 데이터 프레임은 여러분들이 고급프로그래밍 첫번째 과제로 실습한 happiness 데이터 입니다.
우리는 happiness 데이터의 df를 happiness 데이터와 동일한 열 순서를 갖도록 만들어 봅시다.

```python
happiness = pd.read_csv("happiness.csv")
happiness.head(5)
```

```python
df.head(5)
```

Pandas는 df.drop method를 통하여 데이터 프레임의 행 또는 열을 제거 할 수 있습니다.

먼저 df에서 Country 열을 제거한 데이터 프레임을 만들어 봅시다.

```python
#DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
df2 = df.drop("Country",axis=1)
df2.head(5)
```

df.drop method는 여러개의 열을 제거하는 것 또한 가능합니다.
df.drop(["a","b"],axis="select")

```python
df3 = df.drop(["Country","Region"],axis=1)
df3.head(5)
```

하지만 이 방법은 열을 지정하는 방법과 마찬가지로 열의 이름이 긴 경우 굉장히 번거롭다는 단점이 있습니다.
따라서 이번에는 아까 배운 df.columns를 사용해 열을 추출해 보도록 하겠습니다. 

```python
drops = df.columns[[0,1,2]]
df4 = df.drop(drops,axis=1)
df4.head(5)
```

두 데이터 프레임이 비슷한 모양을 가졌습니다. 두 데이터 프레임들 사이에 가장 큰 차이점읜 "Happiness Score"의 위치의 차이가 될 것입니다.
이번에는 "Happiness Score"를 데이터 프레임의 맨마지막 열로 변경해 봅시다.

```python
df4.head(5)
```

```python
happiness.head(5)
```

```python
Happy = df4.columns[0]
HS = df4[Happy]
HS.head(5)
```

```python
df5= df4.drop(Happy,axis=1)
df5.head(5)
```

```python
import copy
Final = copy.deepcopy(df5) #깊은복사를 위함
Final[Happy] = HS
```

```python
Final.head(5)
```

만들어진 DataFrame을 파일로 만들어 봅시다.
pandas는 데이터프레임을 여러 파일 포맷으로 저장할수 있도록 함수를 제공해 주고 있습니다.
우리는 그 중 to_csv와 to_excel을 사용해 봅시다.

DataFrame.to_csv(path_or_buf=None, sep=', ', na_rep='', float_format=None, columns=None, header=True, index=True, index_label=None, mode='w', encoding=None, compression='infer', quoting=None, quotechar='"', line_terminator=None, chunksize=None, tupleize_cols=None, date_format=None, doublequote=True, escapechar=None, decimal='.')

```python
Final.to_csv("_happiness.csv",index=False)
```
